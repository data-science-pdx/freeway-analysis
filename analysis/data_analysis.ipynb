## 3 Preview the data

```
dataset = pd.read_csv("/content/data/FreewayData/freeway_loopdata1hr.csv")
```

* **detectorid**: Id of the detector (key)
* **starttime**: Start time of the time interval represented by this data (key)

* **volume**: 1-hour volume for this detector
* **speed**: 1-hour average speed for this detector (mph)
* **occupancy**: 1-hour average occupancy for this detector (percent)
* **countreadings**: the number of readings received from this detector during the time interval
* **vmt**: 1-hour vehicle miles travelled for this detector (miles)
* **vht**: 1-hour vehicle hours travelled for this detector (hours)
* **delay**: 1-hour delay for this detector (minutes)
* **traveltime**: 1-hour travel time for this detector (minutes)

### 3.3. Missing Data in Bar Chart

```
total = dataset.isnull().sum()
percent = dataset.isnull().sum()/dataset.isnull().count()

missing_data = pd.concat([total,percent], axis=1, keys=['total', 'percent'])
ax = plt.subplots(figsize=(12, 3))
sns.barplot(x = missing_data.index, y = missing_data['percent'])
plt.xlabel('Features')
plt.ylabel('Percent of missing values')
plt.title('Percent missing data by feature')
plt.show()
```

### 3.4 Using Missingno approach
https://github.com/ResidentMario/missingno

#### I. Bar Chart
The `msno.bar` is a simple visualization of nullity by column

```
msno.bar(dataset, figsize=(12,6))
plt.title('Missing data by feature')
plt.show()
```

#### II. Heatmap
The `msno.heatmap` measures nullity correlation: how strongly the presence or absence of one variable affects the presence of another

```
msno.heatmap(dataset,figsize=(10,6))
```

#### III. Dendrogram
The `msno.dendrogram` allows you to more fully correlate variable completion, revealing trends deeper than the pairwise ones visible in the correlation heatmap

```
msno.dendrogram(dataset, figsize=(10,3))
```

#### IV. Matrix
The `msno.matrix` nullity matrix is a data-dense display which lets you quickly visually pick out patterns in data completion.

```
msno.matrix(dataset, figsize=(10,6))
```

### 4 Data cleaning

### 4.1 Replacing With Mean

```
cleaning_flag = True
```

```
if cleaning_flag:
  dataset["speed"] = dataset["speed"].fillna(dataset["speed"].mean())
  dataset["vmt"] = dataset["vmt"].fillna(dataset["vmt"].mean())
  dataset["vht"] = dataset["vht"].fillna(dataset["vht"].mean())
  dataset["delay"] = dataset["delay"].fillna(0.0)
  dataset["traveltime"] = dataset["traveltime"].fillna(dataset["traveltime"].mean())
```

### 4.2 Check if their is not any missing value

```
msno.bar(dataset, figsize=(10,3), color="skyblue")
plt.show()
```

## 5 Exploratory data analysis

### 5.1 Speed Distribution

```
dataset.sort_values(by=['detectorid'], inplace=True)
```

```
x = np.array_split(range(len(dataset)), 4)
for i in x:
  print()
  ax = plt.subplots(figsize=(12, 3))
  sns.barplot(x = "detectorid", y = "speed", data=dataset[i[0]:i[-1]])
  plt.title("speed Distribution dataset["+str(i[0])+":"+str(i[-1])+"]")
  plt.ylabel("speed")
  plt.xlabel("detectorid")
  plt.show()
```

### 5.2 Volume Distribution

```
x = np.array_split(range(len(dataset)), 4)
for i in x:
  ax = plt.subplots(figsize=(12, 3))
  sns.barplot(x = "detectorid", y = "volume", data=dataset[i[0]:i[-1]])
  plt.title("Volume Distribution of dataset["+str(i[0])+":"+str(i[-1])+"]")
  plt.ylabel("volume")
  plt.xlabel("detectorid")
  plt.show()
```
